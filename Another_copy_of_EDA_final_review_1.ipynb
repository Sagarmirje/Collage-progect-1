# **Comparision between PHCS and CHCs statewise using Grouped Bar **

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
file_path = "/content/Final_EDA_datasheet(1).csv"  # Update with correct path if needed
df = pd.read_csv(file_path)

# Selecting required columns
states = df["State/ UT - Col. 1"]
chcs = df["Number of CHCs Functioning - Col. 2"]
phcs = df["Number of PHCs Functioning"]

# Creating a grouped bar chart
x = np.arange(len(states))  # the label locations
width = 0.4  # the width of the bars

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width/2, chcs, width, label='CHCs', color='blue')
rects2 = ax.bar(x + width/2, phcs, width, label='PHCs', color='green')

# Adding labels, title, and custom x-axis labels
ax.set_xlabel("States/UTs")
ax.set_ylabel("Number of CHCs and PHCs")
ax.set_title("Comparison of CHCs and PHCs across States")
ax.set_xticks(x)
ax.set_xticklabels(states, rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```

Inferance:States with a higher number of CHCs might have better healthcare coverage.
If a state has a low number of CHCs but a high population, it may indicate inadequate healthcare services.

```python
# @title Filling NA value with Linear Rigression Methode
columns_with_na = [
    'Number of Community Health Centres - With functioning Stabilization Units for New Born - Col. 8',
    'Number of Community Health Centres - Functioning as per IPHS norms - Col. 16',
    'No. of CHC having a regular supply of - AYUSH drugs for common ailments - Col. 18',
    'No. of PHCs Functioning as per IPHS norms'
]
from sklearn.linear_model import LinearRegression
import pandas as pd

# Make sure Population has no NaNs
df1 = df1[df1['Population'].notna()]

# Loop through each target column
for col in columns_with_na:
    if df1[col].isna().sum() > 0:
        # Prepare training data (non-null rows)
        train_data = df1[df1[col].notna()]
        X_train = train_data[['Population']]
        y_train = train_data[col]

        # Train Linear Regression model
        model = LinearRegression()
        model.fit(X_train, y_train)

        # Predict missing values
        missing_data = df1[df1[col].isna()]
        X_pred = missing_data[['Population']]
        y_pred = model.predict(X_pred)

        # Fill NaNs with predicted values
        df1.loc[df1[col].isna(), col] = y_pred

        # Optional: round if it's a count column
        df1[col] = df1[col].round().astype('Int64')

```
